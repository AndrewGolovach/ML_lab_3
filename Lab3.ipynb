### Задание 1.
Необходимо вычислить $VCdim(H)$, если $H$ - семейство линейных бинарных классификаторов в d-мерном пространстве.

Предположим, что $VCdim(H)=d$.

 Пусть $C = \{(1,0,...,0), (0,1,...,0), ..., (0,0,...,1)\}$. Рассмотрим произвольный элемент из $H_c$: $$y^* = \{y_1, y_2, ..., y_d\}, y_i \in \{-1,+1\}$$ $$h^* = sign(<w,x>+b) = sign(\sum((1,1,...,1) * x_k) - 0.5) \in H$$ 
 где $x_k$ - элемент, для которого $y_k = +1$. Объединив полученные гипотезы, получим семейство, раскрашивающее $C$.
 
Попробуем раскрасить $C$, состоящее из $d+1$ элемента. Элементы $C$ лежат в $d$-мерном пространстве, следовательно, любой элемент из $C$ можно представить в виде линейной комбинации остальных элементов. Такой элемент, при применении $h \in H$, даст однозначное значение $y$, следовательно, $C$ размера $d+1$ не раскрашивается $H$.

### Задание 3.

Пусть $X$ - булев гиперкуб размерности n. Для множества $I \in {1, 2,..., n}$ и объекта $x \in X$, $x = (x_1, x_2, ..., x_n)$ зададим функцию $h_I(x) = \left( \sum x_i \right) mod 2$. Требуется найти $VCdim$ множества всех таких функций.

Предположим, $VCdim(H) = n.$ 

Рассмотрим $C = \{(1,0,...,0), (0,1,...,0), ..., (0,0,...,1)\}$ и $y^* = {y_1, y_2, ..., y_d}, y_i \in \{0,1\}$, $y \in Y$.
Выберем $i$ такие, что $y_i = 1$, сформируем множество $I_C = \{i: y_i = 1\}$. Строим по этому множество $h_{I_C}(x)$, которое, в силу выбора $C$, $h_C(x)$ примет вид: $$h_{I_C}(x) = \begin{cases} 1, & \mbox{если порядковый номер x в множестве $C \in I_C$} \\ 0, & \mbox{иначе} \end{cases}$$ Объединия полученные гипотезы, получим семейство, раскрашивающее $C$.

Любое множество $C$ размерности $n+1$ нельзя раскрасить, так как $$VCdim(H)≤log2|X|=n$$.
